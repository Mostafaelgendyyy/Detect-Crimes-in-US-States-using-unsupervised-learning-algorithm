# K-Means Clustering Algorithm

```python
DatasetNP=Dataset_Copy.drop(columns=['Unnamed: 0','UrbanPopulation'])
DatasetNP=np.array(DatasetNP)
DatasetNP
```

```python
NumberCluster=int(input("Enter Number of Clusters: "))
```

```python
AllFinalDistance=[]
```

```python
class KMeans:
    def __init__(self,NumberCluster):
        self.NumberOfClusters=NumberCluster
        self.Centroids=self.GetRandomCentroids(NumberCluster)
        self.Results=[]
    def GetRandomCentroids(self,NumOfClusters):
        index = np.random.choice(len(DatasetNP), NumOfClusters, replace=False)
        centroids = []
        for i in index:
            centroids.append(DatasetNP[i])
        return centroids
    def dis_Manhattan(self,Data_Point): # get distance between test point and Centroid
        Distances=[]
        for pnt in self.Centroids:
            Manhattan=np.sum(abs(Data_Point-pnt))
            Distances.append(Manhattan)
        return Distances
    def getResults(self):
        return self.Results # used for Getting OUTLIERS
    def fit(self):
        PreviousClusters=[]
        ResultedClusters=[]
        Iterations=0
        while True :
            Iterations+=1
            print(Iterations)
            PreviousClusters=ResultedClusters.copy()
            ResultedClusters.clear()
            for DataPoint in DatasetNP:
                DistanceFromCentroids=self.dis_Manhattan(DataPoint)
                MinDistance=np.min(DistanceFromCentroids)
                IndexMinDistance=DistanceFromCentroids.index(MinDistance)
                AllFinalDistance.append(MinDistance) # for Outliers
                ResultedClusters.append(IndexMinDistance+1)
            if (PreviousClusters == ResultedClusters):
                self.Results = PreviousClusters.copy()
                break
            AllFinalDistance.clear()
            self.Centroids=[]
            for index in range(self.NumberOfClusters):
                Cluster=index+1
                ClusterPoints=[]
                for i in range(len(ResultedClusters)):
                    if (ResultedClusters[i] == Cluster):
                        ClusterPoints.append(DatasetNP[i])
                List=[]
                for Mean in np.mean(ClusterPoints,axis=0):
                    List.append(Mean)
                self.Centroids.append(List)
        Dataset_Copy['Cluster']=PreviousClusters
```

```python
kmeans= KMeans(NumberCluster)
kmeans.fit()
```

```python
################# Required Output in Assignment
ResultedClustersOutput=kmeans.getResults()
States=[]
for i in range(NumberCluster):
    ClusterStates=[]
    for j in range(len(ResultedClustersOutput)):
        if ResultedClustersOutput[j]==i+1:
            ClusterStates.append(Dataset_Copy['Unnamed: 0'].iloc[j])
    States.append(ClusterStates)
for i in range(len(States)):
    print('Cluster: ',i+1)
    print(States[i])
```

# Outliers

### METHOD 1: Getting Outliers Using Quartiles

```python
AllFinalDistance
```

```python
Outliers=[]
```

```python
AllClusteredFinalDistance=[]
for i in range(NumberCluster):
    ClusterDistancePoints=[]
    for j in range(len(AllFinalDistance)):
        if ResultedClustersOutput[j] == i+1:
            ClusterDistancePoints.append(AllFinalDistance[j])
    AllClusteredFinalDistance.append(ClusterDistancePoints)
            
AllClusteredFinalDistance
```

```python
OutliersOfEachCluster=[]
for Cluster in AllClusteredFinalDistance:
    Q1,Q3 = np.percentile(Cluster,[25,75])
    IQR=Q3-Q1
    MaxDISTANCEBOUND= Q3+(1.5*IQR)
    MinDISTANCEBOUND= Q3-(1.5*IQR)
    print(MaxDISTANCEBOUND)
    for Distance in Cluster:
        if Distance > MaxDISTANCEBOUND or Distance < MinDISTANCEBOUND:
            OutliersOfEachCluster.append(Distance)
```

```python
for Distance in OutliersOfEachCluster:
    index = AllFinalDistance.index(Distance)
    Outliers.append(Dataset_Copy.iloc[index])
```

```python
Outliers=pd.DataFrame(Outliers)
Outliers 
```
